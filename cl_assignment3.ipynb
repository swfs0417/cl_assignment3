```python
from dotenv import load_dotenv
from openai import OpenAI
load_dotenv()
```

```python
print(response.choices[0].message.content)
```

```python
dataset = {
  "의문문": "{topic}에 대해 알려줄 수 있어?",
  "명령문": "{topic}에 대해 자세히 설명해 줘.",
  "평서문": "{topic}를 알고 싶어.",
  "청유문": "우리 {topic}에 대해 알아보자.",
  "정중한 요청": "{topic}에 대해 알려주실 수 있으신가요?",
  "의문문(간접)": "혹시 {topic}에 대해 알고 있니?",
  "감탄문": "{topic} 신기하다!",
  "확인 요청": "{topic}이 사실이야? 자세히 말해줘.",
  "간단 요청": "{topic}에 대해 알려줘.",
  "미완결": "{topic}이 널리 알려져 있는데, 그 이유는",
  "주제": "{topic}",
  "요약": "{topic} 관련된 정보",
}
```

```python
topics = [
  '허준이 교수의 신라면 사랑',
  '휘문고 총기난사 사건',
  '기아 K5 액션빔 튜닝',
  '서울대학교 기숙사 증축사업',
  '2025년 총선',
  '수아당 서울대입구점',
  '스리랑카뿔버섯',
  '롯데리아 게살버거',
  '요루시카 내한공연 후기',
  '고종커피 신메뉴'
]
```

```python
print(*topics, sep='\n')
```

```python
dataset
```

```python
result = {topic: {} for topic in topics}

for topic in topics:
    for key, value in dataset.items():
        client = OpenAI()
        response = client.chat.completions.create(
            model="gpt-4o-mini-2024-07-18",
            messages=[
                {
                    "role": 'user',
                    "content": value.format(topic=topic)
                }
            ],
            max_tokens=1000, temperature=0.4
        )
        result[topic][key] = response.choices[0].message.content
```

```python
result
```

```python
import json
result_json = json.dumps(result, ensure_ascii=False, indent=2)
with open('result.json', 'w', encoding='utf-8') as f:
    f.write(result_json)
```
