```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

applyfont = "Malgun Gothic"
plt.rc("font", family=applyfont)
plt.rc("axes", unicode_minus=False)
```

```python
df = pd.read_csv('result.csv', encoding='cp949', index_col=0)
```

```python
df.sort_values(by='평균', axis=1,  ascending=False, inplace=True)
```

```python
df
```

```python
colors = ['#FF9999' if x > 0 else '#66B3FF' for x in df.loc['평균', :]]
colors = ['#5DFF78' if df.columns[i] == '평균' else colors[i] for i in range(len(df.columns))]
```

```python
# barplot 그리기 (표준편차를 오차막대로 표시)
sns.barplot(data=df, errorbar = 'sd', palette=colors)
plt.xticks(np.arange(0, len(df.columns)), rotation=45)
plt.title('문장 형태별 평균 hallucination 점수')
plt.show()
```
